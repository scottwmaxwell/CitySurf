```python
# Create boundingbox from boundingboxes values
boundingbox = []
for boundingbox_coords in df['boundingboxes']:
    box = {
        "type": "Polygon",
        "coordinates":[[
            [boundingbox_coords[2], boundingbox_coords[0]],
            [boundingbox_coords[2], boundingbox_coords[1]],
            [boundingbox_coords[3], boundingbox_coords[1]],
            [boundingbox_coords[3], boundingbox_coords[0]],
            [boundingbox_coords[2], boundingbox_coords[0]]
        ]] # For some reason a double array is needed
    }
    boundingbox.append(box)


df['boundingbox'] = boundingbox

# drop boundingbox
df = df.drop('boundingboxes', axis=1)
```
